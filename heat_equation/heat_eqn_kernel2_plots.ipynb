## Heat Eqn: exploring parameter space

#### NLML function without log scaled parameters

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('heat_data.pkl', 'rb') as input:
    heat_data = pickle.load(input)
```

```python
np.min(heat_data['nlml']), np.max(heat_data['nlml'])
```

```python
plt.figure()
CS = plt.contour(heat_data['theta1'], heat_data['theta2'], heat_data['nlml'], np.array([-1e12, -1e11, -1e10, -1e7, -1e6, -1e5]))
plt.clabel(CS, inline=1, fontsize=3)
plt.show()
```
